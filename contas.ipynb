Aula01 -> PYTHON CONTAS

```python
a = 10
b = 3
c = 2.75

print("a + b =", a + b)
print("a - c =", a - c)
print("b * c =", b * c)

print("\na / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)

print("\na / c =", a / c)
print("a // c =", a // c)
print("a % c =", a % c)
```

*   // -> Operador que pega o valor **inteiro** da divisão
*   % -> Operador que pega o módulo da divisão **(resto da divisão)**

```python
# Estudando operação Resto %

print("a % b", a % b)
print("a % c", a % c)
```
