```python
#Cálculo do IMC:

#ENTRADAS
altura = float(input('Forneça sua altura em metros: '))
peso = float(input('Forneça seu peso em kilogramas: '))

#PROCESSAMENTO
imc = peso / (altura * altura)
imc_arredondado = round(imc, 2)

#SAÍDA
print('\nSeu IMC é:', imc_arredondado)
```

Round(valor, 2) serve para definir a quantidade de casas decimais de uma variável.

```python
#Ler a altura e a base de um triângulo e calcular sua área = base * altura / 2:

#ENTRADAS
altura = float(input('Forneça a altura do triângulo em metros: '))
base = float(input('Forneça a base do triângulo em metros: '))

#PROCESSAMENTO
area = (base * altura) / 2
area_arredondada = round(area, 2)

#SAÍDAS
print('\nA área do triângulo é igual a:', area_arredondada, 'metros')
```

```python
#A velocidade 100km/h corresponde a 62mph -> Ler uma velocidade em KM/H e converter para MPH

#ENTRADA
velocidade_kmh = float(input('Forneça uma velocidade em km/h: '))

#PROCESSAMENTO
velocidade_mph = velocidade_kmh * 0.62
velocidade_mph_arrendondada = round(velocidade_mph, 2)

#SAÍDA
print('\nA velocidade fornecida em mph é igual a:', velocidade_mph_arrendondada)
```

```python
#Descobrir quantas horas e minutos em tenho em 17043 segundos -> depois crie um input para entrada

#ENTRADA
tempo_segundos = int(input('Forneça um tempo em segundos: '))

#PROCESSAMENTO
tempo_hora = tempo_segundos // 3600
tempo_hora_resto = tempo_segundos % 3600

tempo_minutos = tempo_hora_resto // 60

#SAÍDA
print('\nEm', tempo_segundos,'segundos eu possuo', tempo_hora,'horas e', tempo_minutos,'minutos')
```

```python
#Programar um caixa eletrônico que possui notas de 100, 50, 20 e 10  -> fazer uma lógica que traga a menor quantidade de notas disponíveis para sacar um determinado valor

#ENTRADA
valor = int(input('Forneça um valor para sacar: '))

#PROCESSAMENTO
nota_100 = valor // 100
nota_50 = (valor % 100) // 50
nota_20 = ((valor % 100) % 50) // 20
nota_10 = (((valor % 100) % 50) % 20) // 10

#SAÍDA
print('\nAo sacar ', valor, ' você deve receber como menor combinação de notas possível:\n\n',nota_100,' nota(s) de cem\n',nota_50,' nota(s) de 50\n',nota_20,' nota(s) de 20\n',nota_10,' nota(s) de 10\n', sep='')
```

Sep=' ' resolve o problema de espaços em branco gerados pelo \n

```python
#Para converter C° em °F fazemos -> F = 9/5C+32 -> Ler temperatura em C° e coverter para °F

#ENTRADA
celcius = int(input('Forneça uma temperatura em graus Celcius: '))

#PROCESSAMENTO
fahrenheit = (9/5) * celcius + 32
fahrenheit_arredondado = round(fahrenheit, 2)

#SAÍDA
print('\n',celcius,' C° em fahrenheit é igual a ', fahrenheit_arredondado,' °F', sep='')
```

```python
# Um motorista de táxi deseja calcular o rendimento de seu carro na praça. Sabendo-se que o preço do combustível é de R$ 4.90, escreva um algoritmo para ler: a marcação do odômetro (Km) no início do dia, a marcação (Km) no final do dia, o número de litros de combustível gasto e o valor total (R$) recebido dos passageiros. Calcular e escrever: a média do consumo em Km/L e o lucro (líquido) do dia.

#ENTRADAS
combustivel = 4.90
km_inicio = float(input('Forneça a marcação do odômetro no início do dia em Km: '))
km_fim = float(input('Forneça a marcação do odômetro no final do dia em Km: '))
litros_de_combustivel_gastos = float(input('Forneça o número de litros de combustível gasto: '))
valor_total = float(input('Forneça o valor total recebido dos passageiros: '))

#PROCESSAMENTO
media_comsumo = (km_fim - km_inicio) / litros_de_combustivel_gastos
media_comsumo_arredondada = round(media_comsumo, 2)

lucro = valor_total - (litros_de_combustivel_gastos * combustivel)
lucro_arredondado = round(lucro, 2)

#SAÍDA
print('\nSua média de consumo foi de', media_comsumo_arredondada, 'litros por quilômetro e seu lucro (líquido) do dia foi de', lucro_arredondado, 'reais')
```
